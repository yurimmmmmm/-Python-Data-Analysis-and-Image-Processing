```python
import matplotlib.pyplot as plt
import cv2

image = cv2.imread('img_A.jpeg', cv2.IMREAD_GRAYSCALE)
ret, result1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
ret, result2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
ret, result3 = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC)
ret, result4 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)
ret, result5 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)

images = []
images.append(result1)
images.append(result2)
images.append(result3)
images.append(result4)
images.append(result5)

for i in images:
  plt.imshow(cv2.cvtColor(i, cv2.COLOR_GRAY2RGB))
  plt.show()
```
