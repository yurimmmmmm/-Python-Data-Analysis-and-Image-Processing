```python
# 이미지의 크기를 조절: cv2.resize(image, dsize, fx, fy, interpolation)
# 가로 비율: fx, 세로 비율: fy
# interpolation - INTER_CUBIC: 사이즈를 크게 할 때, INTER_AREA: 사이즈를 작게 할 때
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('img_A.jpeg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

image1 = cv2.imread('img_A.jpeg')
expand = cv2.resize(image1, None, fx=2.0, fy=2.0, interpolation=cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(expand, cv2.COLOR_BGR2RGB))
plt.show()

image2 = cv2.imread('img_A.jpeg')
shrink = cv2.resize(image2, None, fx=0.6, fy=0.6, interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(shrink, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# cv2.warpAffine(image, M, dsize): 이미지의 위치를 변경합니다
# M: 변환 행렬
# disze: Manual Size
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('img_A.jpeg')

# 행과 열 정보만 저장합니다.
height, width = image.shape[:2]

M1 = np.float32([[1, 0, 500], [0, 1, 100]])
dst = cv2.warpAffine(image, M1, (width, height))
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# cv2.getRotationMatrix2D(center, angle, scale): 이미지 회전을 위한 변환 행렬
# center: 회전중심
# angle: 회전 각도
# scale: Scale Factor
```

```python
image = cv2.imread('img_A.jpeg')

# 행과 열 정보만 저장합니다.
height, width = image.shape[:2]

M = cv2.getRotationMatrix2D((width / 2, height / 2), 90, 0.3)
M1 = cv2.getRotationMatrix2D((width / 2, height / 2), 90, 0.5)
dst = cv2.warpAffine(image, M, (width, height))
dst1 = cv2.warpAffine(image, M1, (width, height))
print(M)
print(M1)

plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(dst1, cv2.COLOR_BGR2RGB))
plt.show()
```
