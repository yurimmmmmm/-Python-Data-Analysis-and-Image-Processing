```python
image = cv2.imread('img_A.jpeg')

# 픽셀 수 및 이미지 크기 확인
print(image.size)
print(image.shape)

# 이미지 Numpy 객체의 특정 픽셀을 가리킨다.
px = image[100,100]

# B, G, R 순서로 출력됨
# 단, GRAY SCALE은 B, G, R로 구분 X
print(px)

# R 값만 출력하기기
print(px[0])
print(px[1])
print(px[2])
```

```python
import matplotlib.pyplot as plt

for i in range(0, 1000):
  for j in range(0, 1000):
    image[i, j] = [255, 255, 255]

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

image[0:1000, 0:1000] = [255, 255, 255]
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_GRAY2RGB))
plt.show()
```

```python
image= cv2.imread('img_A.jpeg')

# Numpy Slicing: ROI 처리 기능
roi = image[1000:1750, 2500:3250]

# ROI 단위로 이미지 복사하기
image2 = image.copy()
image[0:750, 0:750] = roi

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.show()
```
